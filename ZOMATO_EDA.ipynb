```python
df_country = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/zomato_eda/Country-Code.xlsx')
```

```python
df_country
```

```python
final_df = pd.merge(df,df_country,on = 'Country Code',how = 'left')
```

```python
final_df.head(2)
```

```python
final_df.columns
```

```python
Country_names = final_df.Country.value_counts().index
Country_values = final_df.Country.value_counts().values
```

```python
plt.pie(Country_values, labels=Country_names,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.show()
```

```python
plt.pie(Country_values[:3], labels=Country_names[0:3],
        autopct='%1.1f%%')
plt.show()
```
